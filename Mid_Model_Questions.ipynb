```python
#2
#recursive function to find the factorial of a number

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

factorial(5)
    
```

```python
#6
#fibanacci series through recursion
def fib(n):
    if n==0 or n==1:
        return n
    return fib(n-1)+fib(n-2)

n = int(input("Enter the number: "))
for i in range(n):
    print(fib(i), end=" ")
```

```python
#8
#palindrome check without string method
num = int(input("Enter a number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10          
    reverse = reverse * 10 + digit  
    num = num // 10           
if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")

```

```python
#10
#GCD of 2 numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b != 0:
    a, b = b, a % b

print("GCD is:", a)

#alternate
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while a != b:
    if a > b:
        a -= b
    else:
        b -= a

print("GCD is:", a)

```

```python
#18
#factorial using recursion
def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)

n = int(input("Enter a number: "))
print("Print factorial of",n,"is",fact(n))
```
